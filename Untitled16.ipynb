Data Cleaning and EDA on Titanic dataset.

Linke of the dataset: https://www.kaggle.com/c/titanic/data?select=train.csv

```python
# Import Required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
data = pd.read_csv("titanic.csv")

# Show first 10 row od dataset
data.head(100)
```

```python
print(data.describe)
```

```python
# Get summary from the dataset

data.describe()
```

```python
# Illustrate a concise summary of a DataFrame.
# Useful for understanding the structure and content of your dataset at a glance.

# Show columns name, data type, null values or missing values and etc.

data.info()
```

```python
# Identify and count the number of missing values (NaNs) in each column of a DataFrame

data.isnull().sum()
```

As we can see there are some non_values or missing values in columns "Age", "Cabin" and "Embarked". We need to deal with these columns by proper methods.

```python
# Handle missing values in a DataFrame

data.dropna(subset=["Embarked"], inplace=True)
data["Cabin"] = data["Cabin"].fillna("Unknown")
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data.isnull().sum()
```

```python
# Identify and count the number of duplicate rows in a DataFrame

data.duplicated().sum()
```

A histogram with a kernel density estimate (KDE) for the "Age" column.

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Plot for Age
plt.figure(figsize=(12, 6))
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

A count plot for the "Sex" column.

```python
# Plot for Sex
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=data)
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

A count plot for the "Survived" column.

```python
# Plot for Survived
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=data)
plt.title('Count of Passengers by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

A count plot for the "Pclass" column.

```python
# Plot for Pclass
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=data)
plt.title('Count of Passengers by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot for Survived by Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Count of Passengers by Sex and Survival')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

The chart shows that more male passengers were on the Titanic than female passengers, but a significantly higher proportion of female passengers survived. This suggests that the "women and children first" protocol was followed during the evacuation, resulting in fewer surviving males despite their larger number. This highlights the social and cultural norms of the time and the challenges faced by men during the disaster.

```python
df = data

# Create Age Classes
def age_class(age):
    if age <= 12:
        return 'Children (0-12)'
    elif 13 <= age <= 19:
        return 'Teenagers (13-19)'
    elif 20 <= age <= 64:
        return 'Adults (20-64)'
    else:
        return 'Elderly (65+)'

# Apply the function to create a new column 'Age Class'
df['Age Class'] = df['Age'].apply(age_class)

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot for Survival by Age Class
plt.figure(figsize=(10, 6))
sns.countplot(x='Age Class', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Age Class')
plt.xlabel('Age Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Not Survived (0)', 'Survived (1)'])
plt.xticks(rotation=45)
plt.show()
```

The chart shows the number of passengers on the Titanic who survived and did not survive, categorized by age class. It reveals that while adults had the highest overall survival rate, children and elderly passengers had significantly lower survival rates. Teenagers had a survival rate that fell between these two groups. This suggests that the "women and children first" protocol was prioritized during the evacuation, benefiting children but not necessarily elderly passengers.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

The scatter plot shows the relationship between age and fare for passengers on the Titanic, with different colors representing whether they survived (1) or not (0). It reveals a general trend where younger passengers tended to pay lower fares, while older passengers, especially those above 50, tended to pay higher fares. However, there is also a significant amount of variation within each age group, suggesting that other factors besides age influenced ticket prices. Additionally, the plot indicates that survival was not solely determined by age or fare, as there are overlapping points for both survivors and non-survivors across different age and fare ranges.
